```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x,y,'r+')
plt.xlabel("x")
plt.ylabel("y")
plt.title("title")
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y)
plt.subplot(1,3,2)
plt.plot(x,y,'r-')
plt.subplot(1,3,3)
plt.plot(x,y,'c')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("x")
```

```python
fig = plt.figure()

axes = fig.add_axes([.1,.1,.8,.8])
axes1 = fig.add_axes([.9,.1,.8,.8])

axes.plot(x,y)
axes1.plot(x,y,'r')
```
